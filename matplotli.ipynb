## Matplotlib API 概觀

### plt.plot (df.plot)
> 繪圖的懶人包（包括 df.plot ），會幫你直接在最近使用的圖表上繪圖，沒有的話就會自動新建一個。
### ax.plot
> 手動建立的 API，提供多項客製化的參數與圖表控制。


## Matplotlib 的核心繪圖觀念
### 不同參數代表的意思
> * Figure：可以把他當作畫布，一張畫布上可以有很多的 Axes。
> * Axes：就是我們俗稱的「子圖」，每個 Axes 一次只能在一張畫布上。

```python
#### import numpy as np 
import matplotlib.pyplot as plt

# 將 figure 實例化到 fig 的變數上，並設定寬與高（此處為預設值）
fig = plt.figure(figsize=(6,4))

# 將 axe 新增到實例化後 2×2 的 figure 上，並指定第 1,2,3,4 個
ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)
```

```python
# 畫出圖表
plt.plot([1,2,3,4])

# 為圖表加上 x 座標以及x, y的label
# 把線換成紅色的點r代表red，o代表圓點
plt.plot([1,2,3,4], [1,4,9,16], "ro")
plt.ylabel("y label")
plt.xlabel("x label")
```

產出數值介於0~5的array，並以0.2為間隔，畫出y=x, y=x², y=x³ 的圖表，r 代表紅色、兩個-代表虛線，b代表藍色、s代表方塊，g代表綠色，^代表三角形

```python
t = np.arange(0.,5.,0.2)
# red dashes, blue square and green trinagle
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^',)
```

```python
#用random.randint來生成一個隨機整數序列
import random
def random_int_list(start, stop, length):
    start, stop = (int(start), int(stop)) if start <= stop else (int(stop), int(start))
    length = int(abs(length)) if length else 0
    random_list = []
    for i in range(length):
        random_list.append(random.randint(start, stop))
    return random_list
```

```python
x = random_int_list(1, 30, 10)
print('x: ', x)
y = random_int_list(1, 30, 10)
print('y: ', y)
plt.bar(x,y)
```

```python
x = random_int_list(1, 45, 20)
print(x)
y = random_int_list(1, 45, 20)
print(y)
plt.scatter(x,y)
```

## Matplot與Pandas的整合
以往我們用matplot畫圖需要將pandas資料丟進matplot畫圖，不過因為現在pandas有整合matplot，所以可以在pandas的DataFrame下直接畫圖

```python
# 本次使用 iris資料集做練習
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names) # 轉換為 data frame
iris_data.plot.scatter(x= 'sepal length (cm)', y='sepal width (cm)')
# 在Dataframe的格式上直接.plot 之後就可以接matplotlib的函式了
```

## plt.style
matplotlib還提供一個style的設定可以直接透過改變style來讓圖形變好看

```python
#使用plt.style.available列出可以用的style有哪些
plt.style.available
```

```python
plt.style.use('ggplot')
iris_data.plot.scatter(x= 'sepal length (cm)', y='sepal width (cm)')
```

```python
plt.style.use('classic')
iris_data.plot.scatter(x= 'sepal length (cm)', y='sepal width (cm)')
```
